## Scikit-learn
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

**Scikit-learn** is the most popular Machine Learning library. 

<img width="250" src="https://user-images.githubusercontent.com/7065401/71116995-003c4080-21b4-11ea-851e-2d87ea6801e3.png" />

[Scikit-learn](https://scikit-learn.org/), or _sklearn_, is an open source machine learning library that supports supervised and unsupervised learning. It also provides various tools for model fitting, data preprocessing, model selection and evaluation, and many other utilities.

It provides a large list of tools and algorithms, divided in 6 categories:

- **Classification**, to identify which category an object belongs to.
- **Regression**, to predict a continuous-valued attribute associated with an object.
- **Clustering**, to perform automatic grouping of similar objects into sets.
- **Dimensionality reduction**, to reduce the number of random variables to consider.
- **Model selection**, to compare, validate and choose parameters and models.
- **Preprocessing**, to extract and normalize features.

All implemented algorithms in scikit-learn, whether preprocessing, supervised learning, or unsupervised learning algorithms, are implemented as classes.

These classes are called **Estimators** in scikit-learn. To apply an algorithm, you first have to instantiate an object of the particular class.

For example, let's see how can we perform a [Linear Regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html):

````python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model
````

The estimator class contains the algorithm, and also stores the model that is learned from data using the algorithm.


##Â Fit method
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

All estimators have a **_fit_** method, which is used to build the model. 

```python
estimator.fit(x_train, [y_train])
```

The `fit` method always requires as its first argument the data `X`, represented as a NumPy array or a SciPy sparse matrix, where each row represents a single data point. The data `X` is always assumed to be a NumPy array or SciPy sparse matrix that has continuous (floating-point) entries.

Supervised algorithms also require a `y` argument, which is a one-dimensional NumPy array containing target values (labels) for regression or classification (i.e., the known output labels or responses).

````python
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100,
                       n_features=1,
                       noise=25)

X.shape, y.shape
````

````python
X[0], y[0]
````

````python
X[1], y[1]
````

````python
import pandas as pd

df = pd.DataFrame({'X': X.ravel(), 'y': y.ravel()})
df.head()
````

````python
df.plot(kind='scatter',
        x='X',
        y='y',
        color='black',
        linewidth=3,
        figsize=(12,6))
````

````python
model.fit(X, y)
````

##Â Predict method
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

There are two main ways to apply a learned model in scikit-learn.

To create a prediction in the form of a new output like $y$, you use the `predict` method.

```python
estimator.predict(X_test)
```

To create a new representation of the input data, you use the `transform` method.

> We will cover this `transform` method in detail on upcoming lessons.

````python
y_pred = model.predict(X)

y_pred
````

````python
X[0], y[0], y_pred[0]
````

````python
X[1], y[1], y_pred[1]
````

````python
df['y_pred'] = y_pred
````

````python
ax = df.plot(kind='scatter',
             x='X',
             y='y',
             color='black',
             linewidth=3,
             figsize=(12,6))

df.plot(kind='line',
        x='X',
        y='y_pred',
        color='red',
        figsize=(12,6),
        ax=ax)
````

##Â Score method
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

Additionally, all supervised models have a `score` method that allows an evaluation of the model.

```python
estimator.score(X_test, y_test)
```

````python
model.score(X, y)
````

Now we will go through a simple machine learning application and create our first model.

We will be using the [Fish market dataset](https://www.kaggle.com/aungpyaeap/fish-market) which contains data about common fish species for fish market.

![fish](https://user-images.githubusercontent.com/7065401/101499773-6a616e80-394c-11eb-824b-532b593fff3f.jpg)

Having this data we want to **estimate the weight of fish based on the other columns**: <u>**a regression problem**</u>.

To begin with, let's load into a pandas DataFrame the `fish.csv` dataset:

````python
df = pd.read_csv('data/fish.csv')

df
````

We will be analizing characteristics of each fish such as length, height and width.

WE have 159 different fishes. Each individual is called **sample** in machine learning, and their properties are called **features**.

The **shape** of the data array is the number of samples multiplied by the number of features.

````python
df.shape
````

## Understanding the data
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

The dataset has the following features:

- **Species**: Name of the fish species
- **Weight**: Weight of fish in gram
- **Length1**: Vertical length in cm
- **Length2**: Diagonal length in cm
- **Length3**: Cross length in cm
- **Height**: Height in cm
- **Width**: Diagonal width in cm

Also we could get an idea of the types of the attributes we have:

````python
df.info()
````

````python
df['Species'].value_counts()
````

The dataset is imbalanced with fishes in different species ranging from 6 to 56. This indicates that predicting the species of the fish is a tough job here as we do not have enough training data. We would rather try to predict the weight of the fishes.

We can look if our data has any null value:

````python
df.isnull().sum()
````

or take a look at a summary of each attribute including average values (mean), minimum, maximum as well as some percentiles (25th, 50th or media and 75th e.g. values at this points if we ordered all the values for an attribute).

````python
df.describe()
````

## Visualizing data and its relationships
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

Before diving into the creation of a machine learning model to predict the weight of a fish, it is often a good idea to inspect the data, to see if the task is easily solvable without machine learning, or if the desired information might not be contained in the data.

Let's take a look at the interactions between the variables:

````python
from pandas.plotting import scatter_matrix

ax = scatter_matrix(df[['Weight', 'Length1', 'Length2', 'Length3', 'Height']],
                    diagonal='kde',
                    figsize=(12,12))
````

````python
import seaborn as sns
````

````python
sns.pairplot(df, hue='Species')
````

## Correlation matrix 
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

A correlation matrix is a table showing correlation coefficients between variables.

Correlation indicates pairwise correlation between continuous variables in the dataset. While calculating the correlation, categorical variables are ignored, so 'Species' has been automatically ignored from our dataset.

````python
df.corr()
````

````python
ax = sns.heatmap(df.corr(), center=0, cmap="coolwarm", annot=True)
ax.set_title(u"Correlation Matrix", fontsize=14)
````

Heatmap is a great way to visually look at the relationship between variables.

We can see that many variables indicate positive correlation with a few variables showing very strong positive correlation with values close of 0.9.

The variables `Length1`, `Length2` and `Length3` are highly correlated leading to multicolinearity. Thus, we can drop two of the length variables and just keep one in order to do our regression analysis.

> We usually drop variables with correlation coefficient, $r = \pm  0.9$ or above.



---
### Analyzing pair of variables

To validate what we saw on our correlation matrix we can use a **pairplot** which allows us to see both the distribution of the variable and pairwise relationship between variables in the dataset.

````python
sns.pairplot(df, hue='Species')
````

The perfect linear representation of graphs between `Length1`, `Length2` and `Length3` indicates that our idea of dropping two from these three is correct.

````python
df.drop(['Length2', 'Length3'], axis=1, inplace=True)
````

````python
df.head()
````

The pairplot indicates that the explanatory variables have linear relationship with the dependent variable, so it is good to fit a linear regression model to this dataset.

## Features and Labels
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)


On the `df` DataFrame we have **features** and **labels** (also known as $X$ and $y$).

---
###Â Features

Features ($X$) are individual independent variables that act as the input in your system. Prediction models use features to make predictions.

---
### Labels

Labels ($y$), also known as **_target variable_**, are the final output. You can also consider the output classes to be the labels.

````python
#all variables except Weight and Species
X = df.iloc[:,2:]

#Weight variable only
y = df['Weight']
````

## Train and Test sets
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)


For that, we need to split our data into a training set and a testing set. **We can't train our model with the same data we'll use to test it later**, that'd be cheating ðŸ˜‰.

<img width="300" src="https://user-images.githubusercontent.com/7065401/70933003-cafbeb00-2019-11ea-99a5-0b7fa8b3068d.png" />

Let's split the dataset into training and testing:

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=123)
````

````python
print("Shape of X_train: ", X_train.shape)
print("Shape of y_train: ", X_test.shape)
print("Shape of X_test: ", y_train.shape)
print("Shape of y_test: ", y_test.shape)
````

## Building our first model
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

We'll use a Linear Regression algorithm to predict `weight`:

````python
lm = LinearRegression(random_state=123)

lm
````

To build the model on the training set, we call the `fit` method of the `model` object, which takes as arguments the NumPy array `X_train` containing the training data and the NumPy array `y_train` of the corresponding training labels:

````python
lm.fit(X_train, y_train)
````

````python
print("Intercept of the model: ", lm.intercept_)
print("Coefficient of the model: ", lm.coef_)
````

````python
print("Y = ", lm.intercept_, "+", lm.coef_[0], "*", df.columns[2], "+", lm.coef_[1], "*", df.columns[3], "+", lm.coef_[2], "*", df.columns[4])
````

Done. We have our first model trained and ready to predict! ðŸš€

## Making predictions
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

Our `X_test` is part of our original dataset, so we can test our model and see how good it perform.

Let's predict the weight of our test fishes:

````python
y_pred = lm.predict(X_test)

y_pred
````

We have our predictions, but what are the real values?

````python
df_y_test = pd.DataFrame(y_test)
df_y_test1 = df_y_test.reset_index(drop=True)

df_y_test1.head()
````

````python
df_y_pred = pd.DataFrame(y_pred, columns=["Predicted weight"])
df_y_pred.head()
````

````python
df_y = pd.concat([df_y_test1, df_y_pred], axis=1)
df_y.head()
````

## Evaluating the Model
![orange-divider](https://user-images.githubusercontent.com/7065401/98619088-44ab6000-22e1-11eb-8f6d-5532e68ab274.png)

We will use $R^2$ to evaluate our model.

$R^2$ (R-squared) evaluates the scatter of the data points around the fitted regression line. It is also called the coefficient of determination, or the coefficient of multiple determination for multiple regression. 

![r2](https://user-images.githubusercontent.com/7065401/101543788-774d8480-3983-11eb-81e1-6bf8d8a84a5b.png)

Usually, the larger the $R^2$, the better the regression model fits your observations.


````python
from sklearn.metrics import r2_score

print(r2_score(y_test, y_pred))
````
